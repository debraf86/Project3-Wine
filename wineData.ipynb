```python
# read the csv files - 1 per year
dfRedWine = pd.read_csv("./winequality-red.csv")
dfRedWine.head()
```

```python
fixedAcidityMax = dfRedWine["fixed acidity"].max()
fixedAcidityMin = dfRedWine["fixed acidity"].min()
fixedAcidityCount = dfRedWine["quality"].count()
fixedAcidityCount
dfRedWine["citric acid"].max() #1.0
dfRedWine["chlorides"].max() #.611
dfRedWine["density"].max() # 1.00369
dfRedWine["sulphates"].max() #2.0
dfRedWine["volatile acidity"].max() #1.58
dfRedWine["fixed acidity"].max() #15.9
```

```python
# iterate through each row and normalize between 0 and 1. 

for i in range(len(dfRedWine)) : 
    df.loc[i, "fixed acidity"] = (dfRedWine.loc[i, "fixed acidity"] - dfRedWine["fixed acidity"].min())/ (dfRedWine["fixed acidity"].max()-dfRedWine["fixed acidity"].min())
    df.loc[i, "volatile acidity"] = (dfRedWine.loc[i, "volatile acidity"] - dfRedWine["volatile acidity"].min())/ (dfRedWine["volatile acidity"].max()-dfRedWine["volatile acidity"].min())
    df.loc[i, "sulphates"] = (dfRedWine.loc[i, "sulphates"] - dfRedWine["sulphates"].min())/ (dfRedWine["sulphates"].max()-dfRedWine["sulphates"].min())
    df.loc[i, "density"] = (dfRedWine.loc[i, "density"] - dfRedWine["density"].min())/ (dfRedWine["density"].max()-dfRedWine["density"].min())
    df.loc[i, "residual sugar"] = (dfRedWine.loc[i, "residual sugar"] - dfRedWine["residual sugar"].min())/ (dfRedWine["residual sugar"].max()-dfRedWine["residual sugar"].min())
    df.loc[i, "free sulfur dioxide"] = (dfRedWine.loc[i, "free sulfur dioxide"] - dfRedWine["free sulfur dioxide"].min())/ (dfRedWine["free sulfur dioxide"].max()-dfRedWine["free sulfur dioxide"].min())
    df.loc[i, "total sulfur dioxide"] = (dfRedWine.loc[i, "total sulfur dioxide"] - dfRedWine["total sulfur dioxide"].min())/ (dfRedWine["total sulfur dioxide"].max()-dfRedWine["total sulfur dioxide"].min())
    df.loc[i, "pH"] = (dfRedWine.loc[i, "pH"] - dfRedWine["pH"].min())/ (dfRedWine["pH"].max()-dfRedWine["pH"].min())
    df.loc[i, "alcohol"] = (dfRedWine.loc[i, "alcohol"] - dfRedWine["alcohol"].min())/ (dfRedWine["alcohol"].max()-dfRedWine["alcohol"].min())
    df.loc[i, "quality"] = (dfRedWine.loc[i, "quality"] - dfRedWine["quality"].min())/ (dfRedWine["quality"].max()-dfRedWine["quality"].min())
```
