```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

wine_df = pd.read_csv('winequalityred.csv')
```

```python
wine_df.describe()
```

```python
sns.set()
wine_df.hist(figsize=(10,10), color='red')
plt.show()
```

## Observation

###### It appears that the scale of features is greatly different and it may negatively impact machine learning algorithms. 
###### Normalization is needed to avoid negative impact to the performance of ML algorithms. 

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=wine_df.drop(columns=['quality']), orient='horizontal', palette='Reds',showfliers=False)
```

## Observation

The boxplot below also shows the difference in the scale of features.
Normalization is needed.

```python
wine_df2 = pd.read_csv('normalizedWineData.csv')
```

```python
wine_df2 = wine_df2.drop(['Unnamed: 0'], axis=1)
```

```python
wine_df2.describe()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=wine_df2.drop(columns=['quality']), orient='horizontal', palette='husl', showfliers=False)
```

## Observation

The distribution looks must better after data normalization


```python
sns.set()
wine_df2.hist(figsize=(10,10), color='red')
plt.show()
```

## Observation

###### Quality shows that there are a lot of mediocre wine between 5 and 7 and very little good and bad wine. 
###### Chlorides and residual sugar are not very well distributed. We may compare ML results with and without them.
###### Besides citric acid, all features show normal distribution. Citric acid fluctuates some.

```python
fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'fixed acidity', data = wine_df2)

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'volatile acidity', data = wine_df2)

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'citric acid', data = wine_df2)

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'residual sugar', data = wine_df2)

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'chlorides', data = wine_df2)

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'free sulfur dioxide', data = wine_df2)

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'total sulfur dioxide', data = wine_df2)

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'density', data = wine_df2)

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'pH', data = wine_df2)

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'sulphates', data = wine_df2)

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'alcohol', data = wine_df2)
```

### Observation
Citric acid, sulphates and alcohol go up as quality goes up.

Volatile acidity, chlorides,density and pH go down as quality goes up.

# Correlation 

```python
colormap = plt.cm.viridis
plt.figure(figsize=(20,20))
plt.title('Correlation', y=1.05, size=15)
sns.heatmap(wine_df2.astype(float).corr(),
            vmin=-1, vmax=1,center=0, square=True, 
            cmap=sns.diverging_palette(250, 20, n=200),
            linecolor='white', annot=True)
```

### Observation

[Citric acid and fixed acidity], [density and fixed acidity] and [free sulfur dioxide and total sulfur dioxide] show some correlation, all 0.67. but overalall features are independent from each other. We may use all of them in our ML models.
